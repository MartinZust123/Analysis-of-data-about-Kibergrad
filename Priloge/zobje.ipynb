## Postavitev domnev

Postavimo ničelno domnevo *$H_0$*, ki trdi:

**Dodajanje vitamina C vpliva na dolžino odentoblastov.**

Nato potavimo še alternativno hipotezo *$H_1$*, ki trdi:

**Dodajanje vitamina C ne vpliva na dolžino odentoblastov.**

Recimo, da se odločimo za stopnjo tveganja $\alpha = 0,05$. Odločiti se moramo, kakšen poskus bomo izvedli. Vzamemo povprečje dolžin odontoblastov vsake kategorije, glede na količino vitamina C. 

```python
prva_kat = df[df.KOLICINA == 0.5]
prva_kat
```

```python
druga_kat = df[df.KOLICINA == 1.0]
druga_kat
```

```python
tretja_kat = df[df.KOLICINA == 2.0]
tretja_kat
```

## ANOVA

```python
var1 = prva_kat["DOLZINA"].var()
var1
```

```python
var2 = druga_kat["DOLZINA"].var()
var2
```

```python
var3 = tretja_kat["DOLZINA"].var()
var3
```

```python
residual_var = var1 + var2 + var3
residual_var
```

```python
m1 = prva_kat["DOLZINA"].mean()
m1
```

```python
m2 = druga_kat["DOLZINA"].mean()
m2
```

```python
m3 = tretja_kat["DOLZINA"].mean()
m3
```

```python
m = df["DOLZINA"].mean()
m
```

```python
var_among = (len(prva_kat)*(m-m1)**2 + len(druga_kat)*(m-m2)**2 + len(tretja_kat)*(m-m3)**2)/4
var_among 
```

```python
razmerje = var_among/residual_var
razmerje
```

Razmerje je spet precej veliko. 

## Linearna regresija

```python
C1 = sum(df["KOLICINA"]*df["DOLZINA"]) - len(df) * df["KOLICINA"].mean() * df["DOLZINA"].mean()
C1
```

```python
C2 = sum(df["KOLICINA"]**2) - len(df) * df["KOLICINA"].mean() ** 2
C2
```

```python
b = C1 / C2
b
```

```python
a = df["DOLZINA"].mean() - b * df["KOLICINA"].mean()
a 
```

```python
import numpy as np

x_os = [0]
y_os = [a]
x = 0
for i in range(200):
    x += 0.01
    x_os.append(x)
    y_os.append(b * x + a)

x_values = np.array(x_os)
y_values = np.array(y_os)

plt.scatter(df["KOLICINA"], df["DOLZINA"])
plt.plot(x_values, y_values,c="r")

plt.show()
```

## IZRAČUN INTERVALA ZAUPANJA 

```python
vsota_kvadratov = sum(df["KOLICINA"]**2)
kvadrat_vsote = sum(df["KOLICINA"])**2
n = len(df)
ulomek = n / (n * vsota_kvadratov - kvadrat_vsote)
varia = sum((df["DOLZINA"] - b*df["KOLICINA"] - a)**2)/(n-2)
SE = varia * ulomek
SE
```
